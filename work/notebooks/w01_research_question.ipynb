## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

What this work can claim: observed patterns in this anonymized sample — for
example, that 54% of pages (16,262 of 30,000) currently show a declining
trend_direction, or that certain content types or freshness tiers correlate
with declining performance. These are directional signals, useful for
decision-support: helping an editor decide where to look first, not telling
them exactly what will happen if they act.

What this work can never claim: causation (that refreshing a page will cause
its traffic to recover), certainty about any individual page's future
performance, or any insight into how Google's search ranking algorithm
actually works. A page marked "down" isn't guaranteed to improve if refreshed,
and a page marked "up" isn't guaranteed to keep climbing. The dataset is also
anonymized and limited to 32 clients, so patterns here are observed in this
sample only — they haven't been validated against real-world outcomes yet.

All results in this notebook should be read as observed / measured /
directional / decision-support only.
