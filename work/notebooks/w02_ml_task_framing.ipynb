## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

Content Refresh lane, framed as a scoring/ranking problem, not classification or
clustering. The goal isn't to sort pages into fixed categories. It's to produce
a continuous "refresh priority" score for every page, then rank pages from most
to least worth an editor's time. Ranking fits because the real decision (which
50 pages does the editor tackle this month) only needs a relative ordering, not
a hard yes/no label.

label — no one has recorded "this page was refreshed and it worked," so the

## 3. Success metric

*One metric you can defend. What number means 'good'?*

a declining trend_direction and non-trivial search_volume (e.g., above the
median). This is the right metric because the real action only touches a
fixed number of pages per month (an editor's capacity), so what matters is
whether the top of the list is genuinely worth the effort — not how accurately
I predict an exact score for every page in the dataset.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule (e.g., "flag every page with trend_direction == down") is too
crude: 16,262 of 30,000 pages (54%) are already "down," far more than any
editor can act on, and a simple rule can't tell which of those matter most.
It also can't combine multiple weak signals (trend magnitude, search volume,
freshness, engagement) into one ranked list, or improve as outcome data comes
in over time. A scoring/ranking approach can weigh these signals together and
be tuned against real editor outcomes — something a static if-statement can't do.
